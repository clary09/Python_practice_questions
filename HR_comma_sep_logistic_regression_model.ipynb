```python
y=df['left']
y
```

```python
col=df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
col
```

```python
dummies = pd.get_dummies(col.salary, prefix="salary")
df_final = pd.concat([col,dummies],axis='columns')
df_final.drop('salary',axis='columns',inplace=True)
df_final
```

```python
x=df_final
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.5,random_state=50)
```

<h3><strong>Model formation:</strong></h3>

```python
from sklearn.linear_model import LogisticRegression
md = LogisticRegression()
md
```

```python
md.fit(x_train,y_train)
```

<h3><strong>Accuracy:</strong></h3>

```python
md.score(x_test,y_test)
```
